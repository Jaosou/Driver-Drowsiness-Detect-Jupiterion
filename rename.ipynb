```python
import os
from os import listdir
import shutil
from os.path import join
import cv2
import matplotlib.pyplot as plt
```

```python
path = 'archive/Dateset/'
target_path_close = 'Closed'
target_path_open = 'Open'
train_path = [path + f for f in listdir(path)]

closed_dir = 'Closed'
opened_dir = 'Open'
os.makedirs(closed_dir,exist_ok=True)
os.makedirs(opened_dir,exist_ok=True)

img_count = 0
```

```python
for item in train_path :
    img_path = [item + '/' + f for f in listdir(item)]
    for item_path in img_path :
        for i,flie_name in enumerate(listdir(item_path)):
            file_path = join(item_path,flie_name)

            if file_path.lower().endswith(('.jpg')) :
                if 'close' in file_path.lower():
                    new_name = f'Close_{img_count}.png'
                    img_count += 1
                    dest_path = join(target_path_close,new_name)
                elif 'open' in file_path.lower() :
                    new_name = f'Open_{img_count}.png'
                    img_count += 1
                    dest_path = join(target_path_open,new_name)
                else :
                    continue
            shutil.copy(file_path, dest_path)
            print(dest_path)
```

```python
path = 'ExImg/Open/'
count =  24000
for i,imgPath in enumerate(listdir(path)) :
    print(imgPath)
    new_name = f"{path}ex_img_{count}.png"
    os.rename((path + imgPath),new_name)
    print(new_name)
    count += 1
```
